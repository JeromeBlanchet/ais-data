```python
import os
import pandas as pd
import numpy as np

import plotly.express as px

from IPython.display import display
```

```python
# Load data
data_dir = "./Data/"

dfs = [pd.read_csv(data_dir + f) for f in
       [file for file in os.listdir(data_dir) if file.endswith(".csv")]]
print(len(dfs))

df = pd.concat(dfs, ignore_index=True)

# Convert to datetime
df['dt_pos_utc'] = pd.to_datetime(df['dt_pos_utc'])
df['mmsi'] = df['mmsi'].astype(str)

display(df)
```

```python
# Check types
df.dtypes
```

```python
# Print unique nav status
print(df['nav_status'].unique())

# Filter for moored ships
df = df[df['nav_status'] == "Moored"]
```

```python
# Create a column for dt_pos_utc_hour and dt_pos_utc_day
df['dt_pos_utc_date'] = df['dt_pos_utc'].dt.date
df['dt_pos_utc_hour'] = df['dt_pos_utc'].dt.hour

# Drop duplicates based on mmsi and date and hour
df_moored_date = df.drop_duplicates(subset=['mmsi', 'dt_pos_utc_date'], ignore_index=True)
df_moored_hour = df.drop_duplicates(subset=['mmsi', 'dt_pos_utc_hour'], ignore_index=True)

df_date = df_moored_date.groupby("dt_pos_utc_date")['mmsi'].count().reset_index()
df_hour = df_moored_hour.groupby(["dt_pos_utc_date", "dt_pos_utc_hour"])['mmsi'].count().reset_index()

# Get number of unique ships per day and hour
display(df_date)
display(df_hour)
```

```python
# Bar Graph: Date by unique MMSI moored
fig = px.bar(df_date, x='dt_pos_utc_date', y='mmsi')
fig.show()
```

```python
port_stats2 = df.groupby(['mmsi', 'vessel_type', 'dt_pos_utc_date', 'dt_pos_utc_hour']).agg({'mmsi': 'count',
                                                                                          'dt_pos_utc': ['min', 'max']}).reset_index()
port_stats2.columns = ['mmsi', 'vessel_type', 'dt_pos_utc_date', 'dt_pos_utc_hour',
                       'obs_count', 'dt_pos_utc_min', 'dt_pos_utc_max']
port_stats2['arrival_date'] = port_stats2.dt_pos_utc_min.dt.date
port_stats2['time_in_port'] = port_stats2.dt_pos_utc_max - port_stats2.dt_pos_utc_min
port_stats2['time_in_port_sec'] = port_stats2['time_in_port'].dt.total_seconds()
port_stats2['time_in_port_min'] = port_stats2['time_in_port_sec'] / 60
port_stats2['time_in_port_hr'] = port_stats2['time_in_port_min'] / 60

# Filter for time_in_sec > 0
port_stats2 = port_stats2[port_stats2['time_in_port_sec'] > 0]
port_stats2.reset_index(drop=True, inplace=True)

# Display
port_stats2
```

```python
# Average time in Port by vessel_type per hour
avg_time_st_hr = port_stats2.groupby(['vessel_type', 'dt_pos_utc_hour'])['time_in_port_sec'].mean()
avg_time_st_hr = pd.DataFrame(avg_time_st_hr).reset_index()
avg_time_st_hr
```

```python
# Bar graph: average time in port in s by hour by type of ship
fig2 = px.bar(avg_time_st_hr, x='dt_pos_utc_hour', y='time_in_port_sec', color='vessel_type')
fig2.show()
```

```python
# Histogram of time spent in hrs at port by frequency of ships
fig3 = px.histogram(port_stats2, x='time_in_port_hr', color='vessel_type', nbins=12)
fig3.show()
```

```python
# Histogram of time spent in hrs at port by percentage of ships
norm_port_stats2 = port_stats2.groupby(['vessel_type', 'dt_pos_utc_hour']).agg({"mmsi": "size"}).reset_index()
norm_port_stats2.columns = ['vessel_type', 'dt_pos_utc_hour', 'ship_count']
display(norm_port_stats2.head())

# Divide by total ships at this hour
ships_by_hr = port_stats2.groupby('dt_pos_utc_hour').agg({"mmsi": "size"}).reset_index()
ships_by_hr.columns = ['dt_pos_utc_hour', 'total_ship_count']
display(ships_by_hr.head())

# Merge with port_stats2
port_stats = norm_port_stats2.merge(ships_by_hr)

# Create percentage
port_stats['norm_ship_count'] = port_stats['ship_count'] / port_stats['total_ship_count']
display(port_stats.head())
```

```python
# Histogram of time spent in hrs at port by frequency of ships
fig4 = px.histogram(port_stats, x='dt_pos_utc_hour', y='norm_ship_count', color='vessel_type')
fig4.show()
```
